```python
import pandas as pd
```

```python
sa2022 = pd.read_excel('/content/drive/MyDrive/Projetos/Bet Tips/A_2022.xlsx')
sa2021 = pd.read_excel('/content/drive/MyDrive/Projetos/Bet Tips/A_2021.xlsx')
sa2020 = pd.read_excel('/content/drive/MyDrive/Projetos/Bet Tips/A_2020.xlsx')
sa2019 = pd.read_excel('/content/drive/MyDrive/Projetos/Bet Tips/A_2019.xlsx')
sa2018 = pd.read_excel('/content/drive/MyDrive/Projetos/Bet Tips/A_2018.xlsx')

df = pd.concat([sa2022, sa2021, sa2020, sa2019, sa2018],ignore_index = True)
```

```python
#a2022.head()
df.shape
```

```python
df.dropna(inplace=True)
df.shape
```

```python
#conversão de tipo para utilizar split
df['col'] = df['Resultado'].astype(str)
df['Resultado_Casa'] = df['col'].str[0:1]
df['Resultado_Visitante'] = df['col'].str[2:3]
df.drop(['col'], axis = 1, inplace = True)
df.head()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df['Resultado_Casa'] = df['Resultado_Casa'].astype(int)
df['Resultado_Visitante'] = df['Resultado_Visitante'].astype(int)


df['Total Gols'] = df['Resultado_Casa'] + df['Resultado_Visitante']
df.head()
```

```python
#criar coluna nova para condicional
df['Time Vencedor'] = 'NaN'
```

```python
#condicional retornando valor de outra coluna
for index, row in df.iterrows():
    if row['Resultado_Casa'] > row['Resultado_Visitante']:
        df.loc[index,'Time Vencedor'] =  str(df.loc[index,'Em casa'])
    elif row['Resultado_Casa'] < row['Resultado_Visitante']:
        df.loc[index, 'Time Vencedor'] = str(df.loc[index, 'Visitante'])
    else:
        df.loc[index, 'Time Vencedor'] = 'Empate'

df.head()
```

```python
df.drop('Sem', axis=1, inplace=True)
df.drop('Dia', axis=1, inplace=True)
df.drop('Data', axis=1, inplace=True)
df.drop('Horário', axis=1, inplace=True)
df.head()
```

```python
media_gols = df['Total Gols'].mean().round(2)
media_gols
```

```python
df['O/U_0.5'] = 'NaN'
df['O/U_1.5'] = 'NaN'
df['O/U_2.5'] = 'NaN'
df['O/U_3.5'] = 'NaN'

for index, row in df.iterrows():
  if row['Total Gols'] < 0.5:
    df.loc[index,'O/U_0.5'] = 'Under'
  elif row['Total Gols'] > 0.5:
    df.loc[index,'O/U_0.5'] = 'Over'

for index, row in df.iterrows():
  if row['Total Gols'] < 1.5:
    df.loc[index,'O/U_1.5'] = 'Under'
  elif row['Total Gols'] > 1.5:
    df.loc[index,'O/U_1.5'] = 'Over'

for index, row in df.iterrows():
  if row['Total Gols'] < 2.5:
    df.loc[index,'O/U_2.5'] = 'Under'
  elif row['Total Gols'] > 2.5:
    df.loc[index,'O/U_2.5'] = 'Over'

for index, row in df.iterrows():
  if row['Total Gols'] < 3.5:
    df.loc[index,'O/U_3.5'] = 'Under'
  elif row['Total Gols'] > 3.5:
    df.loc[index,'O/U_3.5'] = 'Over'

df
```
